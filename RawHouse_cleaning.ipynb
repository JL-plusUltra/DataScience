```python
####csv and math packs###
import pandas as pd
import numpy as np
import random
from scipy.stats import iqr 
from scipy import stats

from geopy.geocoders import Nominatim
import pandas as pd

import folium
from folium.plugins import FastMarkerCluster

# importing necessary libraries
#import reverse_geocoder as rg
import pprint
 

#visual packs#
import matplotlib.pyplot as plt
import seaborn as sns


################Import Data#########################
data= pd.read_csv('/content/Raw_House_Data.csv')
```

```python
data = data.iloc[:, 1: ]
```

```python
#create z scores to elmininate outliers
data['z_score_sold']=stats.zscore(data['sold_price'])
data['z_score_bedroom']=stats.zscore(data['bedrooms'])
data['z_score_bathroom']=stats.zscore(data['bathrooms'])
data['z_score_sqft']=stats.zscore(data['sqrt_ft'])
data['z_score_fireplace']=stats.zscore(data['fireplaces'])
```

```python
#eliminate outliers beyond three standard deviations
data = data[(np.abs(stats.zscore(data['sold_price'])) < 3)]
data = data[(np.abs(stats.zscore(data['bathrooms'])) < 3)]
data = data[(np.abs(stats.zscore(data['bedrooms'])) < 3)]
data = data[(np.abs(stats.zscore(data['fireplaces'])) < 3)]
data = data[(np.abs(stats.zscore(data['sqrt_ft'])) < 3)]

```

```python
lats=data['latitude'].to_list()
lons=data['longitude'].to_list()
# Creating a zip with latitudes and longitudes
coords=list(zip(lats,lons))
```

```python
len(coords)
```

```python
#build addresses for a map
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="test_app")  # enter a name for your app
full_address=[]
for i in range(len(coords)):
    location = geolocator.reverse(coords[i])
    address=location.raw['display_name']
    full_address.append(address)
```

```python
#add address to dataframe
add=pd.DataFrame(data=full_address , columns=['Address'])
add.head(10)
```

```python
#concatonate address data to our house data
data['address'] = full_address
```
