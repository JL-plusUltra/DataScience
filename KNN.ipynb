```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
def accuracy(y,y_hat):
  return np.mean(y==y_hat)
```

```python
D=2
C=3
N=int(C*1e3)

X0= np.random.randn((N//C),D) + np.array([2,2])
X1= np.random.randn((N//C),D) + np.array([0,-2])
X2= np.random.randn((N//C),D) + np.array([-2,2])

X= np.vstack((X0,X1,X2))
y= np.array([0]*((N//C))+[1]*(N//C)+[2]*(N//C))

```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1], c=y, alpha=.5, s=6)
```

```python
class KNNClassifier():
  def fit(self, X, y):
    self.X=X
    self.y=y

  def predict(self, X, K, epsilon=1e-3):
    N=len(X)
    y_hat= np.zeros(N)

    for i in range(N):
      dist2=np.sum((self.X-X[i])**2, axis=1)
      idxt= np.argsort(dist2)[:K]
      gamma_k=1/(np.sqrt(dist2[idxt]+epsilon))
      y_hat[i] = np.bincount(self.y[idxt], weights=gamma_k).argmax()

    return y_hat
```

```python
np.argmax(np.bincount([0,1,1,1],[(1/2),(1/7),(1/8),(1/9)]))
```

```python
knn = KNNClassifier()
```

```python
knn.fit(X,y)
```

```python
y_hat= knn.predict(X,50)
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1], c=y, alpha=.5, s=6)
accuracy(y,y_hat)
```

```python
X_vis=np.random.uniform(-6,6,(int(N*10),D))
```

```python
y_hat_vis=knn.predict(X_vis,150)
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1], c = y, alpha =.005)
plt.scatter(X_vis[:,0],X_vis[:,1], c= y_hat_vis, alpha=0.03)
```
